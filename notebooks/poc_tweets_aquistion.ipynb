```python
from pymongo import MongoClient
import pandas as pd
import tweepy
import dotenv
import json
import os
```

```python
dotenv.load_dotenv(verbose=True)


CONSUMER_KEY = os.getenv("CONSUMER_KEY")
CONSUMER_SECRET = os.getenv("CONSUMER_SECRET")
ACCESS_TOKEN = os.getenv("ACCESS_TOKEN")
ACCESS_TOKEN_SECRET = os.getenv("ACCESS_TOKEN_SECRET")
```

```python
MONGO_HOST = "mongodb://localhost/twitterdb"
client = MongoClient(MONGO_HOST)
db = client.twitterdb
```

```python
SEARCH_SEEDS = pd.read_excel("../data/input_search.xlsx")
WORDS = SEARCH_SEEDS.key_expression.values.tolist()
```

```python
auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
for query in WORDS:
    print(f"query:{query}", f"progress:{WORDS.index(query)+1}/{len(WORDS)}")
    search_params = {"lang": "pt-br", "tweet_mode": "extended", "q": query + " -filter:retweets"}

    raw_data = tweepy.Cursor(api.search_tweets, **search_params).items(10)

    for status in raw_data:
        json_dump = json.dumps(status._json)
        json_data = json.loads(json_dump)
        json_data["query"] = query
        db.tweets_raw.insert_one(json_data)
```
